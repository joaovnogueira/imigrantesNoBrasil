## Importando o arquivo .csv fonte dos dados: https://servicos.dpf.gov.br/dadosabertos/SISMIGRA/

```python
import pandas as pd
import glob

files = glob.glob("sisMigra.csv")
df_list = []

for f in files:
    df = pd.read_csv(f, encoding='utf8', sep=';')  
    df_list.append(df)

df_all = pd.concat(df_list, ignore_index=True)

df_all.head()
```

## Quantidade de imigrantes por estado no período analisado de 07-2025

```python
df_uf = df_all.groupby('UF')['QTD'].sum().reset_index()
df_uf
```

```python
df_uf = df_uf.sort_values(by='QTD', ascending=False)
df_uf
```

## Descobrindo os países que possuem maior quantidade de imigrantes no Brasil

```python
df_nacionalidade = df_all.groupby('NACIONALIDADE')['QTD'].sum().reset_index()
df_nacionalidade.sort_values(by='QTD', ascending=False)
```

## Descobrindo qual nacionalidade é mais comum por estado

```python
df_nacionalidade_uf = df_all.groupby(['UF', 'NACIONALIDADE'])['QTD'].sum().reset_index()
df_nacionalidade_uf 
```

```python
df_top = df_nacionalidade_uf.loc[df_nacionalidade_uf.groupby('UF')['QTD'].idxmax()].reset_index(drop=True)
df_top
```

## Dentre os mais comuns por estado, qual é os que mais se repetem?

```python
df_top_count = df_top['NACIONALIDADE'].value_counts().reset_index()
df_top_count
```
